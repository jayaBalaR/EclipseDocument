```python
import re
def remove_tagged(text):
    # Remove tagged items Requirement 6
    '''
    Remove the html tags
    '''
    matches = re.findall('<(.*?)>', text)
    for m in matches:
         text=text.replace('<'+m+'>','')
    return text
```

```python
def remove_schars(text):
    for c in [ '\u200c', '-', '\\', '\u200d','[',']','\u200b']: # add newline later
        text=text.replace(c,' ')
    return text
```

```python
names_list = ['words','postag','morphs']
data1 = pd.read_csv('kannada.sample.out_mydata.txt', header=0, names=names_list, sep="\s+")
data1.head(-20)
```

```python
data_dictionary = {}
postaglist= data1['postag'].tolist()
wordslist = data1['words'].tolist()

for word, pos in zip(wordslist,postaglist):
  data_dictionary.update({word:pos})
```

```python
data_dictionary
```

```python
verb_inflection_dict = {}
adj_inflection_dict = {}
pronoun_inflection_dict = {}
adverb_inflection_dict = {}
noun_inflection_dict = {}
for key , value in data_dictionary.items():
    val_str = str(value)
    vstr = val_str.split('.')[0]
    if vstr in ['VM','VB','VBG','VBD','VBN']:
        verb_inflection_dict.update({key:vstr})
    elif vstr in ['RB','RBR','RBS']:
        adverb_inflection_dict.update({key:vstr})
    elif vstr in ['PRP']:
        pronoun_inflection_dict.update({key:vstr})
    elif vstr in ['NN','NNP','NST']:
        noun_inflection_dict.update({key:vstr})
    elif vstr in ['JJ','JJR','JJS']:
        adj_inflection_dict.update({key:vstr})
```
